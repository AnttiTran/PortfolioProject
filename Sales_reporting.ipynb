```python
# Task 1: Import packages and load data

import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path = '/Users/anttitran/Desktop/Data/'
dataframe = pd.read_csv(path + 'sales2019_6.csv')
dataframe
```

```python
# Task 2: Clean and Preprocess data

#Merge 12-month data
filepaths = []
 
for file in os.listdir(path):
    if file.endswith('csv'):
        filepath = path + file 
        filepaths.append(filepath)


df1 = pd.read_csv(filepaths[2])
df1



```

```python
frames = []
for file in os.listdir(path):
    if file.endswith('csv'):
        filepath = path + file 
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)

df = result
df.to_csv('annualSales2019.csv')

```

```python
df.head()
```

```python
# Add 'month' column 

df['Month'] = df['Order Date'].str[0:2]
df.head()
```

```python
df['Order Date'].str[0:2]
```

```python
print(set(df['Month']))
```

```python
# Get rid of 'Nan' or 'Or'

df = df.dropna(how='all')
df.head()
```

```python
df[df['Month'] != 'Or']
df.head()
```

```python
# Task 3: Reporting 
# What was the best month for sales? how much was that month?

df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast='integer', errors='coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast='float', errors='coerce')

df.dropna(subset=['Quantity Ordered', 'Price Each'], inplace=True)

print(df)

# Final check of the DataFrame
print("Cleaned DataFrame:")
print(df)

```

```python
print(df['Quantity Ordered'].dtypes)
print(df['Price Each'].dtypes)
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

```python
moving_column = df.pop('Sales')
df.insert(4, 'Sales', moving_column)
df.head()
```

```python
df.groupby('Month').sum()['Sales']
```

```python
sales_value = df.groupby('Month').sum()['Sales']
sales_value.max()
```

```python
months = range (1,13)
plt.bar(x=months,height=sales_value)
plt.xticks(months)
plt.xlabel('month')
plt.ylabel('Sales in USD')
plt.show()
```
